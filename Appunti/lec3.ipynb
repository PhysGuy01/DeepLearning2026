# Deep Learning - Lezione 3
## Non-linear models

Non e' spessso conveninte lavorare con modelli lineari per certe applicazioni. Se ad esempio volessimo lavorare con immagini 28x28 avrei bisogno di 784 features quindi un tensore enorme e computazionalmente molto dispendioso.

Prendendo spunto dalle connessioni neuronali biologiche, posso pensare di creare un modello matematico di neurone che prende in input esterno dei valori $x_1,\;x_2,\;x_3$ che vengono sommati tra di loro con una somma pesata dove i pesi sono: $w_1,\;w_2,\;w_3$ con eventualmente un certo bias. \
Questa somma passa poi attraverso una funzione di attivazione $g$ che può essere non lineare.

Posso fare una serie di layers di neuroni collegati tra loro. L'esempio in figura mostra una rete "3, 3, 1" cioé 3 valori di input, 3 nel primo hidden layer, 1 output.

Si dicono:
- Feedforward neural network: non c'e' connessione interna e ciclica tra i neuroni dello stesso layer.\
- MLP: feedforward neural netowrk con almeno 3 layers\
- Deep neural network: neural network con più di un hidden layer.\


Per aggiornare i pesi in una neural network si usa la backpropagation.\
Come si fa:
1. Si calcolano tutti i valori dei neuroni tramite le somme pesate (con dei pesi pressoché casuali all'inizio)\
2. Si calcola un errore di predizione per ogni nodo.\
3. Si calcola il gradiente della cost function tramite gli errori e i valori dei neuroni \
4. Si aggionrnano i pesi con un gradient descent sul gradiente della cost function.\

E' importante tenere a mente certe cose:
- Cercare di non inzializzare mai tutti i pesi a 0 perché e' difficile per la rete uscire da quello stato, quindi e' megliao evitare a meno che non sia necessario.
- Inizializzare i pesi a random a volte e' utule per rompere la simmetria dei parametri.
- Inializzare ogni peso su una piccola gaussiana cosi non sono totalmente a caso. 
- Inoltre se abbiamo una sigmoide o una tanh, vogliamo evitare di trovarci lontano dalla parte lineare della funzione. Lontano dalla saturazione.
