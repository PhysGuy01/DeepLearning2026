## Metodologie pratiche

Ci sono molti modi di procedere per migliorare le performance del modello.

### Esempio 1: tecniche di early stopping

Se ho una rete grande che tende ad andare in overfitting, allora posso usare il validation set per valutare se sta davvero migliorando quando si abbassa la loss o se sta overfittando. 

Posso poi andare a prendere i parametri dall'epoca in cui era migliore

### Esempio 2: drop-out

Vado a pesare diverse connessioni tra diversi nodi e uccide i neuroni di troppo, tenendo solo le connessioni necessarie.

## Considerazioni sui dati

Bisogna sempre considerare i dati ed eventuali correlazione tra i parametri. Nella figura di vede una correlazione tra i vari 4 parametri, quindi mi fa supporre che non siano dipendenti tra loro e di poterne tenere anche solo uno tra due scelti (parametro 4 vs parametro 3). 

Quello che fa la vera differenza e' il training e la quantità di risorse disponibili. Importa chi ha più hardware per trovare il modello migliore.

E' molto facile avere dei dataset sbilanciati. In un problema di classificazione potremmo avere ad esempio delle classi più rappresentate o sotto rappresentate. \
Bisogna quindi normalizzare i campioni ('class weighting'), aggiungere nuovi dati artificialmente (ruotare le immagini, distorcerle...) ('oversampling'), ('data augmentation') trasformazioni del dataset senza creare nuovi sample

## Fine-tuning degli iperparametri

Nei violin plot, posso anche prendere un singolo punto di loss molto basso ma non e' detto che sia il migliore in assoluto, perché se cambio il seed potrebbe darsi che non sia ancora il migliore. Quello che mi interessa e' prendere una configurazione stabile (esempio con la coda più corta, adadelta)

- Grid search: Uno dei più banali, abbiamo una griglia (può darsi molto molto grossa -svantaggio-) e proviamo tutti i valori nella griglia uno per uno

- Random search: Simile al grid search ma punti scelti a caso quindi possibilità di prendere il picco più alto.

Questi possono funziuonare abbastanza bene ma ci conviene cercare di capire come si comporta la loss function in funzione degli iperparametri:

- Sequential model based optimization (SMBO): Approssimo la funzione ad una funzione surrogata più facile da maneggiare 

- Tree structured Parzen Estimator (TPE): un tipo si SMBO dove la surrogata e' un modello probabilistico calcolato con la statistica bayesiana.

## Cross-Validation

Spesso e' difficile decidere come dividere il dataset in training/validation/test. Ma esistono degli algoritmi di cross-validation.

Spesso conviene fare un landscape degli iperparametri e poi darla alla TPE piuttosto che lasciar fare tutto alla TPE. Non so che cazzo c'entra ma l'ha detto adesso.

## Closure Testing

a volte può capitare, in fisica, di potermi generare da solo dei dati (?GAN?) in questo caso posso fare un closure test sapendo la legge di base. Modificando la underlying law posso anche trovare i limiti della nostra scelta di parametri.

Questo e' utile se non voglio dipendere troppo dai dati, e posso anche stimare le incertezze.
