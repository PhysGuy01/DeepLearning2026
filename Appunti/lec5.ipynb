### Struttura delle recurrent nn

L'hidden layer di una RNN può essere rappresentata da una funzione 
$$ h^{(t)}=f(h^{(t-1)}, x^{(t)}, w) $$

Ora posso fare passare punti misurati a tempi diversi nel vettore di $x^{(t)}$

Di base esistono 3 architetture possibili
1. x->h(->h<-)->o: ho una rete in cui l'info viene propagata ad un set di layer e mi crea un output correlato. conesione ciclica dentro hidden
2. x->h->o(o->h): gli output di una predizione vengono usati come input della prossima predizione dell'hidden layer. output rientra nell;hidden
3. Una sola predizione di output finale. nessuna connessione ciclica e nesusn output fino all'ultimo

... segue una smerdata di calcoli per la backpropagation ...

Un altro tipo di architetttura del genere sono le architetture encoder/decoder sequence to sequence, in sostanza cerco di creare una memoria passando output negli input futuri.

Questi modelli però non sono né facili da implementare né troppo più potenti rispeto a un MLP. Finché non aumento di molto i parametri almeno.

Esistono anche le Recursive NN (RvNN) che sono più semplici e simili al MLP.
