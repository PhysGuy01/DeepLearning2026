## Algoritmi e tecniche per l'apprendimento dei parametri

Vogliamo minimizzare la loss function $J(w)$, ovvero trovare il minimo (idealmente globale) della funzione.

### Normal equations
Regressione lineare multilineare
Immaginiamo di avere m esempi di training e n features in una matrice X di dimensionalità (m, n). Dobbiamo quindi risolvere il sistema lineare: $Xw=y$
Quando abbiamo un sistema overdeterminato (molte feature credo?):
1. definiamo la cost function : $J(w)=||y-Xw||^2$
2. imponiamo $\frac{\partial J(w)}{\partial w} = 0$
3. Troviamo la soluzione $w = (X^T X)^{-1} X^T y$ 

Questo e' molto complesso computazionalmente.

### Ottimizzazione differenziale
Possiamo basarci sulla soluzione delle derivate per una ottimizzazione genereale
- Algoritmi di ottimizazzione al prim'ordine: gradient descent algorithms. Però non riusciremo a distinguiere un minimo da un massmiimo, al prim'ordine
- Algoritmi al second'ordine: possiamo includere un'hessiana per determinare i minimi, che potrebbe essere migliore della grad desc ma rischia di essere computazionalmente pesante.

Per questo grad descent di solito e' preferibile ma non per tutti i casi.

Immaginiamo di voler minimizzare $J(w)$ dove $w$ e' un vettore di parametri.

- selezioniamo un valore iniziale del vettore $w$ 
- cambiamo w per cercare di ridurre $J(w)$

Ripetere iterativamente finché troviamo un minimo. 
Tuttavia ci sono un po' di problemi:
- gli step vanno calcolati (lo step deve essere dinamico perché rischio di saltar fuori dal minimo o non raggiungerlo mai)
-  ?? non l'ho capita

Possiamo quindi definire gli step di parametri $w$ come:
$w_i = w_i - \eta \frac{\partial J(w)}{\partial w_i}$

Rischia però di rimanere bloccata in una situazione non ideale ma ultimamente sono usciti dei lavori in cui si complementa questa equazione con altri parametri.

$\eta$ e' un altro iperaparametro che richiede fine-tuning:
- $\eta$ piccola ci mette troppo a convergere
- $\eta$ grande rischia di saltare fuori e non convergere affatto

E' importante avere le features $x_i$ in scale comparabili. Ovvero vogliamo standardizzarli e normalizzarli. (Feature scaling) 

Altro problema: se abbiamo un dataset troppo grande dobbiamo tenere i dati in memoria e se non basta dobbiamo andare in fallback e storare i dati sul disco ma questo diventa enormemente inefficiente.

Per risolvere questo problema possiamo dividere i dati in batch:
- batch gradient descent (se vogliamo usare il full batch perché abbiamo pochi dati va specificato, di norma lo fanno automaticamente pytorch e tf)

- stochastic gradient descent: mischia randomicamente esampi di training e usa un esempio per ogni iterazione che può essere utile per esplorare meglio la superficie della loss function, infatti se abbiamo regioni di rumore con minimi locali, questo può essere utile per saltare fuori. Lo svantaggio e' che diventa quasi una random walk e a volte può essere inefficace.

- mini batch gradient descent: unisce i due mondi della BGD e SGD, la più usata normalnente. Di solito usano batch da 32, 64 etc..

- momentum: ricorda $\Delta w$ ad ogni iterazione e aggiorna i parametri tramite l'equazione $w = w - \eta \nabla_w J(w) + \alpha \Delta w$. Molto utile per esplorare la superficie e uscire dai minimi locali.

- RMS propagation: adaptive learning rate per ogni parametro 

- Da qui altri come Adam che e' uno dei più usati.

Si possono anche pensare algoritmi di second'ordine:
 
- Metodo di Newton: $J_T (w) \approx J(w_n) + J'(w_n) \Delta w +\frac{1}{2} J''(w_n)\Delta w^2$. In cui dovremo calcolare però un'hessiana.

- Quasi-newton: utilizza un'approssimazione della matrice hessiana.

### Metodi evoluzionari

In modo un po' ortogonale a questa ricerca di ottimizzazione, si possono usare degli algoritmi basati sull'evoluzione analogamente all'evoluzione biologica.

Un algoritmo genetico e' un esempio di una delle tecniche più popolari di un algoritmo evoluzionario (EA).

E' molto utile quando non abbiamo gradienti o abbiamo funzioni parametriche o non abbiamo una funzione di costo omogenea sul training set

Sostanzialmente abbiamo un punto nello spazio dei parametri, creiamo un tot di cloni e applichiamo delle operazioni trovando dei mutanti, se troviamo una soluzione migliore dell'originale ci spostiamo lì. Quindi non peggioriamo mai, al massimo miglioriamo.
