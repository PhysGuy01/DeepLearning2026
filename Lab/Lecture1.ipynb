Implemento una classe figlia ```Normal``` che restituisce dei valori distribuiti gaussianamente intorno a 0 con varianza 1.

````python
from random import gauss

class Variable:
    def __init__(self, name):
        self.name = name

    def sample(self, size):
        raise NotImplementedError()
    
class Normal(Variable):
    def __init__(self, name):
        super().__init__(name)
    
    def sample(self, size):
        l = []
        for i in range(size):
            l.append(gauss())
        
        return l

Normal("Gauss").sample(3)    
````

## Esercizio 7 - Classe 2

Classe che valuta un polinomio 1-dimensionale con le API:
- costruttore prende il grado come argomento

- metodi `set_parameters` e `get_parameters` per aggiornare la lista di parametri

- contiene metodo `execute` per calcolare il polinomio al punto `x`

````python
class Polynomial:
    def __init__(self, degree):
        self.degree = degree
        self.params = [0]*(degree+1)
            
    def set_parameters(self, position, parameter):
        self.params[position] = parameter

    def get_parameters(self, position):
        return self.params[position]
    
    def execute(self, x):
        y = 0
        for i in range(self.degree + 1):
            y += self.params[i]*(x**i)

        return y

pol = Polynomial(2)
pol.set_parameters(0, -1)
pol.set_parameters(1, 0)
pol.set_parameters(2, 1)

pol.execute(1)
````
