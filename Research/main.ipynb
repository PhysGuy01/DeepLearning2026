```python

import pandas as pd

data = pd.read_parquet('https://huggingface.co/datasets/juliensimon/gravitational-lenses/resolve/main/data/gravitational_lenses.parquet')

print(f'Numero totale di effetti di lensing: {len(data)}')
print(f'Numero di lensing confermati: {len(data[data.grading =='confident'])}')
print(f'Numero di lensing probabili: {len(data[data.grading =='probable'])}')
print(f'Quoziente (conf/prob): {len(data[data.grading =='confident']) / len(data[data.grading =='probable'])}')
```
