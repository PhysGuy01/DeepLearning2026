```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.patches import Patch

barWidth = 0.25

train = [0.779, 0.856, 0.998]
val = [0.717, 0.842, 0.933]
test = [0.715, 0.745, 0.825]

x = np.arange(len(train))
br1 = x
br2 = x + barWidth
br3 = x + 2 * barWidth


train_colors = ['#FF3D3D', "#4d9131", '#ec853c']
val_colors = [ '#FF6161', '#64CD37', '#FFBF6B']
test_colors = ['#FF8585', '#9DDF81','#FFCA85']

fig, ax = plt.subplots(figsize=(12, 8), facecolor='#fff9ef')

plt.rcParams['font.size'] = 15
plt.rcParams['axes.labelsize'] = 'large'

ax.grid(True, linestyle='-.', alpha=0.5, zorder=-100)
ax.set_axisbelow(True)
ax.set_facecolor(color='#fff9ef')

b1 = ax.bar(br1, train, width=barWidth, color=train_colors, edgecolor='grey')
b2 = ax.bar(br2, val, width=barWidth, color=val_colors, edgecolor='grey', )
b3 = ax.bar(br3, test, width=barWidth, color=test_colors, edgecolor='grey')

# Add numeric labels on top of each bar
for bars in (b1, b2, b3):
    for bar in bars:
        height = bar.get_height()
        ax.annotate(str(round(height, 3)),
                    xy=(bar.get_x() + bar.get_width() / 2, height),
                    xytext=(0, 3),
                    textcoords='offset points',
                    ha='center', va='bottom', fontsize=15)

ax.set_xlabel('Models', fontsize=20)
ax.set_ylabel('Accuracy', fontsize=20)
ax.set_xticks(x + barWidth)
ax.set_xticklabels(['Acmonides', 'Brontes', 'Arges'], fontsize=15)

# Create a custom grayscale legend: darkest = train, lightest = test
legend_handles = [
    Patch(facecolor='#333333', edgecolor='grey', label='train'),
    Patch(facecolor='#888888', edgecolor='grey', label='val'),
    Patch(facecolor='#DDDDDD', edgecolor='grey', label='test'),
]
ax.legend(handles=legend_handles, title='Split', loc='upper left')

plt.tight_layout()
plt.savefig('models_comparison.png', dpi = 300)

plt.show()
```
